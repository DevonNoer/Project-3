```python
#establishing a connection to postgres database
connection = pg.connect("dbname=spotify_db user=postgres password=postgres")
#Reading a query to get a table from postgres into pandas
track_df = psql.read_sql_query("SELECT * FROM track", connection)
track_df
```

```python
# Top 10 tracks by streams
top_10_tracks = track_df.sort_values(by='streams', ascending=False).head(10)
top_10_tracks
```

```python
plt.figure(figsize=(10, 8))

top_10_bar_chart = top_10_tracks.plot(kind='bar', x='track', y='streams', color='lightgreen', legend=None)
plt.xlabel('Track')
plt.ylabel('Streams (billions)')
plt.title('Top 10 Tracks by Streams')

plt.xticks(rotation=90)
plt.show()
```

```python
# Merge music_char_df with track_df on the 'track_id' column
combined_df = pd.merge(music_char_df, track_df[['track_id', 'track', 'artists']], on='track_id', how='left')

# Show the top 10 songs by liveness with their track names
top_10_songs_by_liveness = combined_df.sort_values(by='liveness_%', ascending=False).head(10)
top_10_songs_by_liveness = top_10_songs_by_liveness[['track_id', 'track', 'artists', 'liveness_%']]
top_10_songs_by_liveness
```

```python
def songs_in_date_range(date_df, track_df, start_date, end_date):
    # Convert start_date and end_date to Timestamp objects
    start_date = pd.Timestamp(start_date)
    end_date = pd.Timestamp(end_date)
    
    # Convert 'date' column in date_df to Timestamp objects
    date_df['date'] = pd.to_datetime(date_df['date'])
    
    # Filter songs released within the specified date range
    date_range = date_df[(date_df['date'] >= start_date) & (date_df['date'] <= end_date)]
    
    # Merge with track_df to include track names and artists
    songs_within_range = pd.merge(date_range, track_df[['track_id', 'track', 'artists']], on='track_id', how='left')

    return songs_within_range

songs_in_date_range(date_df, track_df, '2021-01-01', '2022-01-01') # Songs released in 2021
```
